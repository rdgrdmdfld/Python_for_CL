#Скрейпим сайт

```python
import requests
from bs4 import BeautifulSoup

url = 'https://api.parliament.uk/historic-hansard/sittings/1930s' # сохраняем ссылку
page = requests.get(url) # загружаем страницу по ссылке
print(page.status_code)  # 200 - страница загружена
```

```python
soup = BeautifulSoup(page.text, 'lxml')
print(soup)
```

```python
for link in soup.find_all('a'): # все гиперссылки на странице
    print(link.get('href'))
```

```python
urls = []

for link in soup.find_all('a'):
  if '19' in link.get('href'):
    urls.append(link.get('href'))

urls
```

```python
full_urls = []

for i in urls:
  u = 'https://api.parliament.uk/historic-hansard/sittings/' + i
  if u not in full_urls:
    full_urls.append(u)

full_urls
```

```python
urls_year = full_urls[4:-1]
urls_year
```

#Дальше перейдем на эти странички и соберем информацию с них

```python
url0 = urls_year[0]
page0 = requests.get(url0)
soup0 = BeautifulSoup(page0.text, 'lxml')
print(soup0)
```

```python
for link in soup0.find_all('a'): # все гиперссылки на странице
    print(link.get('href'))
```

```python
months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
urls = []

for link in soup0.find_all('a'):
  if link.get('href')[-3:] in months:
    urls.append(link.get('href'))

urls
```

```python
full_urls = []

for i in urls:
  u = 'https://api.parliament.uk' + i
  if u not in full_urls:
    full_urls.append(u)

full_urls
```

```python
# выше - ссылки на месяцы 1 года - соберем на все месяцы каждого года
months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']


urls_year_months = []
for i in range(len(urls_year)):
  url0 = urls_year[i]
  page0 = requests.get(url0)
  soup0 = BeautifulSoup(page0.text, 'lxml')
  urls = []
  for link in soup0.find_all('a'):
    if link.get('href')[-3:] in months:
      urls.append(link.get('href'))
  for i in urls:
    u = 'https://api.parliament.uk' + i
    if u not in urls_year_months:
      urls_year_months.append(u)

len(urls_year_months)
```

```python
urls = []

for link in soup0.find_all('a'):
  try:
    if link.get('href')[-2] == '/' or int(link.get('href')[-2:]) < 32:
      urls.append(link.get('href'))
  except:
    pass
urls
```

```python
full_urls = []

for i in urls:
  u = 'https://api.parliament.uk' + i
  if u not in full_urls:
    full_urls.append(u)

full_urls
```

```python
# соберем все ссылки на все дни
urls_day = []
for i in range(len(urls_year_months)):
  url0 = urls_year_months[i]
  page0 = requests.get(url0)
  soup0 = BeautifulSoup(page0.text, 'lxml')
  urls = []
  for link in soup0.find_all('a'):
    try:
      if link.get('href')[-2] == '/' or int(link.get('href')[-2:]) < 32:
        urls.append(link.get('href'))
    except:
      pass
  for i in urls:
    u = 'https://api.parliament.uk' + i
    if u not in urls_day:
      urls_day.append(u)

len(urls_day)
```

```python
urls_day_res = []
```

```python
# увы... нас блокируют
import time
# count_u = 0 при первом запуске, затем меняем
count_u = 0


# соберем все ссылки на все дни

while 'https://api.parliament.uk/historic-hansard/sittings/1939/dec/14' not in urls_day_res:
  urls_day = []
  for i in range(len(urls_year_months[count_u:])):
    url0 = urls_year_months[i]
    page0 = requests.get(url0, headers={'User-Agent': UserAgent().chrome})
    soup0 = BeautifulSoup(page0.text, 'lxml')
    urls = []
    for link in soup0.find_all('a'):
      try:
        if link.get('href')[-2] == '/' or int(link.get('href')[-2:]) < 32:
          urls.append(link.get('href'))
      except:
        pass
    for i in urls:
      u = 'https://api.parliament.uk' + i
      if u not in urls_day and u not in urls_day_res:
        urls_day.append(u)
        time.sleep(3)
  urls_day_res = urls_day_res + urls_day
  n = urls_day_res[-1].rfind('/')
  count_u = urls_year_months.index(urls_day_res[-1][:n])
  print(urls_day_res[-1][:n], count_u, sep=', ')


len(urls_day_res)
```

# Скрейпинг

```python
def GetInf(urls):
  page0 = requests.get(urls)
  soup0 = BeautifulSoup(page0.text, 'lxml')
  section = soup0.find_all('cite', {'class' : 'section'})
  section_res = section[0].text
  text_tag = soup0.find_all('p', {'class' : 'procedural'})
  text = []
  for i in text_tag:
    text.append(i.text.strip())
  result = ''.join(text)
  return section_res, result[2:]
```

```python
result = []

for i in full_urls:
  res = GetInf(i)
  result.append(res)

# (!!!) Важно
# для чужих сайтов используйте функцию sleep из модуля time (не все любят скрейперов!)
```
